```
text= dataset.to_string()
text
```

```
tokenized_text = sent_tokenize(text)
print(tokenized_text)
```

```
tokenized_word=word_tokenize(text)
print(tokenized_word)
print(len(tokenized_word))
```

```
fdist = FreqDist(tokenized_word)
print(fdist)
```

```
fdist.most_common(2)
```

```
v=fdist.plot(75,cumulative=False)
plt.show()
```

# 3 Data Analysis Tasks (ICP)

## Task 1: Removing all stopwords

```
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(text)

filtered_sentence = [w for w in word_tokens if not w in stop_words]
print(filtered_sentence)
print("Number of Words: ", len(filtered_sentence))
```

## Task 3: Plotting the word cloud

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt
     

wordcloud = WordCloud(width = 700, height = 700, 
                background_color ='white', 
                min_font_size = 5).generate(text) 
  
# plotting the WordCloud                         
plt.figure(figsize = (5, 5), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```
